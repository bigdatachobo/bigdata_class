### Split the text into sentences and tokenize the sentences and count the number of words. Draw the bar plot.
### 먼저 문장으로 쪼개고 다시 문장을 단어로 쪼개서 문장별 단어의 갯수를 카운트하고 막대그래프 그리기

```python
token_sent = sent_tokenize(text)
token_sent
```

```python
len_list = []
for i in range(len(token_sent)): #리스트를 숫자로 만들어서 루프
    len_list.append(len(word_tokenize(token_sent[i])))
len_list
```

```python
len_list2 = []
for i in token_sent: #리스트를 하나씩 가져와서 루프
    len_list2.append(len(word_tokenize(i)))
len_list2
```

```python
import numpy as np
plt.bar(np.arange(1,4), len_list2) #각문장에서의 단어의 갯수
plt.show()
```

# 한번에 전처리하는 방법

```python
import re
text="Hello Mr. Smith, how are you doing today? \
The weather is great, and city is awesome.\
The sky is pinkish-blue. You shouldn't eat cardboard."
```

```python
def clean_text(text):
    text = "".join([word.lower() for word in text if word not in string.punctuation])
    tokens = re.split('\W+', text)
    text = [ps.stem(word) for word in tokens if word not in set(stopwords.words('english'))]
    return text
```

```python
clean_text(text)
```

# web scraping example

```python
import requests
from bs4 import BeautifulSoup
```

```python
page = requests.get('https://en.wikipedia.org/wiki/List_of_state_and_union_territory_capitals_in_India')
soup = BeautifulSoup(page.content, 'html.parser')
soup.title.string
```

```python
soup.title.text
```

```python
soup.table
```

```python
all_tables = soup.find_all('table')
```

```python
t = soup.find('table', class_='wikitable sortable plainrowheaders')
t
```

```python
A=[]
B=[]
C=[]
D=[]
E=[]
F=[]
G=[]
```

```python
for row in t.find_all('tr'):
    cells = row.find_all('td')
    #states = row.rind_all('th')
    cells
    if len(cells) == 6: # 빈칸인것은 제외하는 구문
        A.append(cells[0].find(text=True))
        #B.append(states[0].text)
        C.append(cells[1].text)
        D.append(cells[2].text)
        E.append(cells[3].text)
        F.append(cells[4].text)
        G.append(cells[5].text)       
df = pd.DataFrame({'Number':A,
                  #'State':B,
                  'Admin Capital':C,
                  'Leg Capital':D,
                  'Jud Capital':E,
                  'Year':F,
                  'Former Capital':G})
df
```

```python
df.to_excel('../data/table.xls',
            index=None)
```
