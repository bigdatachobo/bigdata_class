```python
import pandas as pd
```

```python
col = pd.read_csv('../data/College.csv')
col.head()
```

## 대학 이름을 index로!

```python
col.set_index(col.columns[0],inplace=True)
col.head()
```

```python
a = col.describe()
a
```

## null 값 찾기

```python
col.isna().sum()
```

# x and y split

```python
y = col.Private
y.head()
```

```python
x = col.drop('Private',axis=1)
x.head()
```

# data preprocessing

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
y = le.fit_transform(y)
y
```

# StandardScaler

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(x)
x = scaler.fit_transform(x)
x
```

```python
print(f'평균: {x.mean()}\n표준편차: {x.std()}')
```

# train and test split

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,
                                                 y,
                                                 test_size = .2)
```

```python
print(f'x_train :\n{x_train}\nx_test: \n{x_test}\ny_train: \n{y_train}\ny_test: \n{y_test}')
```

# 12 nodes with relu, 8 nodes with relu, output with

```python
from keras.models import Sequential
from keras.layers import Dense
```

```python
model = Sequential()
model.add(Dense(12, 
                activation = 'relu',
                input_dim = 17,
                init = 'uniform'))
```

```python
model.add(Dense(8, 
                activation='relu',
                init='uniform'))
```

```python
model.add(Dense(1, 
                activation='sigmoid',
                init='uniform'))
```

```python
model.compile(loss = 'binary_crossentropy',
              optimizer = 'adam',
              metrics=['accuracy'])
```

```python
import keras
```

```python
from keras.callbacks import ModelCheckpoint
```

```python
ckpt_model="./model/weights.best.hdf5"
checkpoint = ModelCheckpoint(ckpt_model,
                             monitor = 'val_acc',
                             verbose=1,
                             save_best_only=True,
                             mode='max')
```

```python
callbacks_list = [checkpoint]
```

```python
history = model.fit(x_train,
                    y_train,
                    epochs = 1000,
                    batch_size = 16,
                    validation_data = (x_test, y_test),
                    callbacks = callbacks_list)
```

```python
y_pred = model.predict_classes(x_test)
```

# evaluation

```python
score = model.evaluate(x_test, y_test)
score
```

```python
met_name = model.metrics_names
met_name
```

```python
print(f'{met_name[0]}: {score[0]}\n{met_name[1]}: {score[1]}')
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred))
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

fig, ax =plt.subplots(1,2,figsize=(17,6))
ax = ax.ravel()

ax[0].set_title('Accuracy Chart')
ax[0].plot(history.history['acc'], label='train')
ax[0].plot(history.history['val_acc'], label='test')
ax[0].legend(loc=4)

ax[1].set_title('Loss Chart')
ax[1].plot(history.history['loss'], label='train')
ax[1].plot(history.history['val_loss'], label='test')
ax[1].legend()
```
