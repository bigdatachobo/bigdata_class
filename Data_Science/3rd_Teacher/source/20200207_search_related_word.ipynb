# 네이버에서 주식데이터 가져오기

```python
url = 'http://finance.naver.com/item/sise_day.nhn?code=005930&page=1'
url
```

```python
import pandas as pd
a = pd.read_html(url, 
                 header=None,
                 na_values=0)[0]
a
```

```python
a.dropna(inplace=True)
a
```

```python
a.reset_index(inplace=True)
a.drop('index',axis=1, inplace=True)
a
```

```python
a.sort_values(by='날짜',ascending=True)
```
