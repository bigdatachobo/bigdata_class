```python
import tensorflow as tf
import numpy as np
```

```python
tf.reset_default_graph()
matrix_1 = np.array([(2,2,2),
                     (2,2,2),
                     (2,2,2)], dtype='int32')

matrix_2 = np.array([(1,1,1),
                     (1,1,1),
                     (1,1,1)], dtype='int32')
matrix_1 = tf.constant(matrix_1)
matrix_2 = tf.constant(matrix_2)
# 행렬곱
matrix_product = tf.matmul(matrix_1, matrix_2) 
# 행렬합
matrix_sum = tf.add(matrix_1, matrix_2)

matrix_3 = np.array([(2,7,2),
                     (1,4,2),
                     (9,0,2)], dtype = 'float32')
# 행렬 판별식
matrix_det = tf.matrix_determinant(matrix_3)

with tf.Session() as sess:
    writer=tf.summary.FileWriter('./graphs', sess.graph)
    result_1 = sess.run(matrix_product)
    result_2 = sess.run(matrix_sum)
    result_3 = sess.run(matrix_det)
print(result_1)    
print(result_2)
print(result_3)
```

# Initialize Variables
### 우선 변수를 초기화시켜서 평가함

```python
a = tf.get_variable(name="A", initializer = tf.constant(2))
b = tf.get_variable(name="B", initializer = tf.constant(3))

c = tf.add(a,b, name='Add')

# add an Op to initialize global variables
init_op = tf.global_variables_initializer()

# launch the graph in a session
with tf.Session() as sess:
    # run the variable initializer operation
    sess.run(init_op)
    # now let's evaluate their value
    print(sess.run(a))
    print(sess.run(b))
    print(sess.run(c))
```

# Exercise #3
### 정규분포로부터 2x3 매트릭스를 만들고, 0이 3개있는 벡터를 만들어 프린트하시요.
#### ▪Weights are usually initialized from a normal distribution using tf.truncated_normal_initializer(stddev=0.01).
#### ▪Biases are usually initialized from zeros using tf.zeros_initializer(). 

```python
weight = tf.get_variable(name="w",
                         shape=[2,3] ,
                         initializer = tf.truncated_normal_initializer(stddev=0.01)) 
bias = tf.get_variable(name="b",
                       shape=[3],
                       initializer = tf.zeros_initializer())
# 초기화
init_op = tf.global_variables_initializer()

with tf.Session() as sess:
    # 초기화
    sess.run(init_op)
    print(f'{sess.run(weight)},\n\n{sess.run(bias)}')
```

# Placeholder Example

```python
a = tf.constant([5,5,5], tf.float32, name = 'A')
b = tf.placeholder(tf.float32, shape=[3], name = 'B')

# 상수벡터와 플레이스홀더를 더함
c = tf.add(a, b, name = 'Add')

with tf.Session() as sess:
    # create a dictionary:
    d = {b:[1,2,3]} # 플레이스홀더 값을 만듦.
    # feed it to the placeholder
    print(sess.run(c, feed_dict=d))
    # 플레이스홀더 값을 이용하여 결과 출력    
```

```python
import numpy as np
import tensorflow as tf

with tf.variable_scope('regression', reuse=tf.AUTO_REUSE):
    X = tf.placeholder(dtype=tf.float32, 
                       shape=[100, 784], 
                       name='X')

    W = tf.get_variable(name="W",
                        shape=[784, 200],
                        initializer = tf.truncated_normal_initializer(mean=0,
                                                                      stddev=0.01),
                        dtype=tf.float32)

    B = tf.get_variable(name="B",
                        shape=[200],
                        initializer = tf.zeros_initializer(),
                        dtype=tf.float32)

    matrix_product = tf.matmul(X, W, name='multiply') 
    matrix_add = tf.add(matrix_product, B, name='add')
    ReLU = tf.nn.relu(matrix_add, name='relu')

    init_op = tf.global_variables_initializer()
```

```python
with tf.Session() as sess:
    sess.run(init_op)
    X_dict = {X:np.random.rand(100,784)}
    sess.run(matrix_product, feed_dict=X_dict)
    sess.run(matrix_add, feed_dict=X_dict)
    print(f'{sess.run(ReLU,feed_dict=X_dict)}')
    writer=tf.summary.FileWriter('./graphs', sess.graph)
```

# 텐서플로우 구현

```python
import numpy as np
import tensorflow as tf
X_h = tf.placeholder(dtype=tf.float32, 
                     shape=[777,17])

W = tf.Variable(tf.random_normal([17,1]))

B = tf.Variable(tf.random_normal([1]))
```

```python
temp = tf.matmul(X_h, W)
temp
```

```python
# y_pred = wx + b
y_pred = tf.add(temp, B)
y_pred
```

```python
loss = tf.reduce_mean(tf.square(y_pred-y))
loss
```

```python
optimizer = tf.train.GradientDescentOptimizer(learning_rate = 0.1)
optimizer
```

```python
train_op = optimizer.minimize(loss)
train_op
```

```python
init_op = tf.global_variables_initializer()
```

```python
with tf.Session() as sess:
    sess.run(init_op)
    for i in range(200):
        sess.run([temp, y_pred, loss, train_op], feed_dict={X_h:x})
        print(sess.run(loss, feed_dict={X_h:x}))
    print(sess.run([W, B]))
```
