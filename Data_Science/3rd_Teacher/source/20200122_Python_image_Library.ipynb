# Labeled Faces in the Wild

```python
import glob
faces = glob.glob('../data/lfw/*/*.jpg')
```

```python
face_list=[]
for face in faces[:100]:
    size = os.path.getsize(face)
    if size != 0:
        temp = Image.open(face).resize((100,100), Image.ANTIALIAS).convert('RGB')
        temp = temp.rotate(130)
        temp = temp.convert('L')
        temp = np.array(temp)
        face_list.append(temp)

face_data = np.array(face_list)
face_data.shape
```

# image mean(), std()

```python
print(f' : {face_data.mean()}\n : {face_data.std()}')
```

# Normalize

```python
face_norm = normalize(face_data,axis=1)
print(f' : {face_norm.mean()}\n : {face_norm.std()}')
```

# 4D array  

```python
face_4D = face_norm.reshape(face_norm.shape[0],face_norm.shape[1],face_norm.shape[2],1)
face_4D.shape
```
