# Air Passenger data preparation

```python
df = pd.DataFrame()
df['t'] = air.iloc[:,0]
df
```

```python
df['t+1'] = df.t.shift(-1)
df['t+2'] = df.t.shift(-2)
df['t+3'] = df.t.shift(-3)
df
```

```python
df.dropna(inplace=True)
df
```

# x and y split

```python
y = df['t+3']
x = df.drop('t+3', axis=1)
```

# CNN model building

```python
from keras.models import Sequential
from keras.layers import Dense
```

```python
model = Sequential()
model.add(Dense(100,
                activation = 'relu',
                input_dim=3))
model.add(Dense(1))
model.compile(optimizer='adam',
              loss='mse')
history = model.fit(x,
                    y,
                    epochs=2000,
                    verbose=1,
                    validation_split=.2)
```

```python
y_pred = model.predict(x)
y_pred
```

```python
plt.figure(figsize=(15,10))
plt.plot(y,label = 'actual')
plt.plot(y_pred, label='predicted')
plt.legend(loc='best')
```

# Evaluation

```python
model.evaluate(x,y)
model.metrics_names
```

```python
h_table = pd.DataFrame(history.history)
h_table
```

```python
plt.figure(figsize=(12,9))
plt.plot(history.history[f'{h_table.columns[0]}'], label=f'{h_table.columns[0]}')
plt.plot(history.history[f'{h_table.columns[1]}'], label=f'{h_table.columns[1]}')
plt.legend()
```
