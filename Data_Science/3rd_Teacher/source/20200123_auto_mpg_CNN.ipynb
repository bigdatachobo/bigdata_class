```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('../data/auto_mpg.csv')
data
```

```python
data.shape
```

```python
data.fillna(0)
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
le.fit(data['car name'])
data['car name'] = le.transform(data['car name'])
data['car name']
```

```python
y=data['mpg']
y
```

```python
x=data.iloc[:,1:]
x
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,
                                                    y,
                                                    test_size=0.2,
                                                    random_state = 42)
```
