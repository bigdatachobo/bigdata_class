# sentiment lexicon

```python
neg = pd.read_csv('../data/opinion-lexicon-English/negative-words.txt',header=None)
neg
```

```python
pos = pd.read_csv('../data/opinion-lexicon-English/positive-words.txt',header=None)
pos
```

```python
neg.columns=['word']
pos.columns=['word']
```

```python
neg['sentiment']='negative'
neg
```

```python
pos['sentiment']='positive'
pos
```

```python
integrated = pd.concat([neg,pos],
                       axis=0,
                       ignore_index=True)
lexicon = integrated.sort_values(by='word',ascending=True)
```

```python
sentiment = lexicon.reset_index()
sentiment
```

```python
sentiment.drop('index',axis=1,inplace=True)
```

```python
sentiment.to_csv('../data/lexicon.csv')
```
