## 6. 학습하기 위해 데이터를 정리하고, 특징을 선택하기

```python
raw_data.head()
```

```python
tmp=[]
for each in raw_data['sex']:
    if each == 'female':
        tmp.append(0)
    elif each == 'male':
        tmp.append(1)
    else:
        tmp.append(np.nan) 
```

```python
raw_data['sex'] = tmp
raw_data.head()
```

```python
raw_data_col=['survived','pclass','sex','sibsp','parch','fare']
for col in raw_data_col:
    raw_data[col] = raw_data[col].astype('float')
raw_data.head(10)    
```

```python
raw_data = raw_data[raw_data['age'].notnull()]
raw_data = raw_data[raw_data['sibsp'].notnull()]
raw_data = raw_data[raw_data['parch'].notnull()]
raw_data = raw_data[raw_data['fare'].notnull()]
raw_data.info()
```

# 7.생존자 예측을 위한 모델 수립

```python
train_pre = raw_data[['pclass','sex','age','sibsp','parch','fare']]
train_pre.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(train_pre,
                                                   raw_data[['survived']],
                                                   test_size=0.1,
                                                   random_state=13)
```

```python
X_train.info()
```

```python
X_train
```

```python
X_train=X_train.reset_index()
X_train=X_train.drop(['index'],axis=1)

X_test=X_test.reset_index()
X_test=X_test.drop(['index'],axis=1)

y_train=y_train.reset_index()
y_train=y_train.drop(['index'],axis=1)

y_test=y_test.reset_index()
y_test=y_test.drop(['index'],axis=1)
```

```python
from sklearn.tree import DecisionTreeClassifier

tree_clf = DecisionTreeClassifier(max_depth=3, random_state=13)
tree_clf.fit(X_train, y_train)

print('Score: {}'.format(tree_clf.score(X_train, y_train)))
```

```python
from sklearn.tree import export_graphviz

export_graphviz(tree_clf,
               out_file="titanic.dot",
               feature_names=['pclass','sex','age','sibsp','parch','fare'],
               class_names=['Unsurvived','Survived'],
               rounded=True,
               filled=True)
import graphviz
with open("titanic.dot") as f:
    dot_graph = f.read()
    
dot=graphviz.Source(dot_graph)
dot.format='png'
dot.render(filename='titanic_tree', directory='../../images/decision_trees', cleanup=True)
```

```python
from sklearn.metrics import accuracy_score

y_pred = tree_clf.predict(X_test)
print("Test Accuracy is", accuracy_score(y_test, y_pred)*100)
```

# 8.디카프리오 생존 예측해보기

```python
# pclass, sex, age, sibsp, parch, fare
dicaprio=[3.,1.,19.,0.,0.,5.] # 위에서 자료형을 float 형식으로 맞췄으므로 그에 맞춰서 숫자+"."(ex-3.)을 붙여준다.
winslet=[1.,0.,17.,1.,1.,100.]
```

```python
tree_clf.predict_proba([winslet])
```

```python
tree_clf.predict([winslet]) # 1. 이면 생존
```

```python
tree_clf.predict([dicaprio]) # 0. 이면 사망
```

```python
def isSurvived(name, person):
    isSurvive = 'not survived' if tree_clf.predict([person])[0] == 0 else 'survived'
    print(name, 'is', isSurvive,
         '-->', max(tree_clf.predict_proba([person])[0])) # 값자체가 [a,b] 형태의 단일 리스트이므로 [0]으로 표시하면 그 안의 리스트 값 
                                                          # 자체만을 나오게한다.
isSurvived('Dicaprio', dicaprio)
isSurvived('Winslet', winslet)
```
