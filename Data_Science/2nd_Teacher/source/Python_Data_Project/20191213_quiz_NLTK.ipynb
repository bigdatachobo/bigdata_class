```
from wordcloud import WordCloud, STOPWORDS
import numpy as np
import pandas as pd
from PIL import Image
import glob
```

# 국문키워드 / 영문키워드 따로 추출 후 csv 파일로 저장 후 open().read() 함수 돌린다.

```
from google.colab import files
uploaded = files.upload()
```

```
a=glob.glob('nltk_quiz_201*')
df=pd.DataFrame()
for exl in a:
    tmp=pd.read_excel(exl)
    df=pd.concat([df,tmp],axis=0)

df_kor=pd.DataFrame()
df_kor=pd.concat([df_kor,df['국문키워드']])

df_eng=pd.DataFrame()
df_eng=pd.concat([df_eng,df['영문키워드']])

df_kor.to_csv('total_exel_kor.txt',encoding='utf-8',index=False,header=False)
df_eng.to_csv('total_exel_eng.txt',encoding='utf-8',index=False,header=False)
```

# key_word_kor

```
text_kor=open('total_exel_kor.txt',encoding='utf-8').read()
key_word_mask=np.array(Image.open('black_book_nltk_quiz.png'))

stopwords = set(STOPWORDS)
stopwords.add(";")  
```

```
plt.figure(figsize=(10,8))
plt.imshow(key_word_mask, cmap=plt.cm.gray, interpolation='bilinear')
plt.axis('off')
plt.show()
```

```
wc = WordCloud(font_path='/usr/share/fonts/truetype/nanum/NanumGothicBold.ttf',
               background_color='white', 
               max_words=2000, 
               mask=key_word_mask, 
               stopwords = stopwords)
wc = wc.generate(text_kor)
wc.words_
```

# key_word_eng

```
text_eng=open('total_exel_eng.txt',encoding='utf-8').read()
key_word_mask=np.array(Image.open('black_book_nltk_quiz.png'))

stopwords = set(STOPWORDS)
stopwords.add(";")  
```

# key_word_eng graph

```
from nltk import regexp_tokenize
pattern=pattern = r'''[a-z]+'''
tokens_en = regexp_tokenize(text_eng,pattern)
tokens_en
```

```
import nltk
eng=nltk.Text(tokens_en, name='English Key Word')
print(len(eng.tokens))
print(len(set(eng.tokens)))
eng.vocab()
```

```
plt.figure(figsize=(12,6))
eng.plot(100)
plt.show()
```
