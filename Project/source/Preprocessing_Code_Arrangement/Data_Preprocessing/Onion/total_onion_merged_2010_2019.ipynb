```python
import pandas as pd
import numpy as np


# 기후 정보
climate=pd.read_csv('../oniondata/adjusted_climate.csv')
```

```python
#대기 정보
air = pd.read_csv('../oniondata/preprocessed_air_merged.csv', encoding='utf-8')
air = air.iloc[:,1:]
air
```

```python
#토양 
soil = pd.read_csv('../oniondata/soil_final_for_merge.csv', encoding='utf-8')
soil
```

```python
# delete column with all 0 values
soil.drop(['시안 CN'], axis=1, inplace=True)
soil.drop(['불소 F'], axis=1, inplace=True)
soil.drop(['유기인'], axis=1, inplace=True)
soil.drop(['유류'], axis=1, inplace=True)
soil.drop(['테트라클로로에틸렌 PCE'], axis=1, inplace=True)
soil.drop(['트리클로로에틸렌 TCE'], axis=1, inplace=True)
soil.drop(['페놀류'], axis=1, inplace=True)
soil.drop(['폴리클로리네이티드비페닐 PCB'], axis=1, inplace=True)
soil
```

```python
onion_unit=pd.read_csv('../oniondata/onion_unit_output.csv',encoding='utf-8')
onion_unit['10a당 생산량 (kg)'].replace('-',np.nan, inplace=True)
onion_unit.dropna(axis=0,inplace=True)
onion_unit
```

```python
element_list = [climate, air, soil]
element=onion_unit
for elem in element_list:
    element=pd.merge(element,
                     elem,
                     on=['year_local','area','year'])
element
```

```python
element.to_csv('../oniondata/onion_columns_all_merge_final.csv', 
               index=False, 
               encoding='utf-8')
```
