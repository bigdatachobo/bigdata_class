```python
makingcost = pd.read_excel('../projectdata/onion_making_cost.xls')
makingcost
```

```python
def year_area(data):
    year = []
    area = []
    for i in data['year']:
        year.append(i[:4])
        area.append(i[5:])
    data['year'] = pd.Series(year)
    data['area'] = pd.Series(area)
```

```python
year_area(makingcost)
makingcost
```

```python
def sido_name(df):
    df['area'] = df['area'].apply(lambda x: '광주광역시' if '광주' in x
                                    else('인천광역시' if '인천' in x
                                    else('부산광역시' if '부산' in x
                                    else('대전광역시' if '대전' in x
                                    else('대구광역시' if '대구' in x
                                    else('서울특별시' if '서울' in x
                                    else('울산광역시' if '울산' in x
                                    else('제주도' if '제주' in x
                                    else x))))))))
```
