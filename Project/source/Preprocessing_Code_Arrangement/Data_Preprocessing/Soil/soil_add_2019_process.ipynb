```python
import pandas as pd
import numpy as np
```

```python
soil = pd.read_csv('../oniondata/soil_proportion.csv', encoding='utf-8')
soil = soil.iloc[:,1:]
soil
```

```python
new_df = pd.DataFrame()
for i in soil.area.unique().tolist():
    new = pd.DataFrame(pd.Series(soil[soil['area'] == i].iloc[-5:].mean()))
    new = new.transpose()
    new_df = pd.concat([new_df,new], axis=0)
new_df
```

```python
new_df['year'] = 2019
new_df['area'] = soil.area.unique().tolist()
new_df
```

```python
year_local = []
for i in range(len(new_df)):
    year_local.append(str(2019) +'_'+ new_df['area'].tolist()[i])
```

```python
new_df['year_local'] = year_local
new_df
```

```python
soil.rename(columns = {'Unnamed: 0.1':'year_local'}, inplace=True)
```

```python
soil_merged = soil.set_index('year_local').append(new_df.set_index('year_local'))
```

```python
soil_merged.to_csv('../oniondata/soil_final.csv')
```
